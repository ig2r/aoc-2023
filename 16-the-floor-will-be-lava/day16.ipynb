```python
import numpy as np

with open('sample.txt', 'r') as f:
  lines = f.read().splitlines()
```

```python
arr = np.asarray(list(map(list, lines)))

directions = [
  (-1,  0),
  ( 1,  0),
  ( 0, -1),
  ( 0,  1),
]

# Items: position (row, col), direction vector (dr, dc)
initial_beams = [
  (np.asarray([0, -1]), np.asarray([0, 1])),
]

def count_energized_tiles(arr, beams):
  energized = np.zeros([*arr.shape, 4])

  while beams:
    beams_prime = []
    #print(f'{len(beams)} beam(s)')

    for (pos, dir) in beams:
      pos_prime = pos + dir

      if not np.all(pos_prime >= np.asarray([0, 0])) or not np.all(pos_prime < arr.shape):
        #print('out of bounds!')
        continue

      dir_index = directions.index(tuple(dir))

      if energized[*pos_prime][dir_index] > 0:
        continue

      energized[*pos_prime][dir_index] = 1
      tile = arr[*pos_prime]

      [dr, dc] = dir
      if tile == '|' and dr == 0:
        # Vertical splitter
        beams_prime.append((pos_prime, np.asarray([-1, 0])))
        beams_prime.append((pos_prime, np.asarray([ 1, 0])))
        continue
      elif tile == '-' and dc == 0:
        # Horizontal splitter
        beams_prime.append((pos_prime, np.asarray([0, -1])))
        beams_prime.append((pos_prime, np.asarray([0, 1])))
        continue
      elif tile == '/':
        # Mirror
        beams_prime.append((pos_prime, np.asarray([-dc, -dr])))
        continue
      elif tile == '\\':
        beams_prime.append((pos_prime, np.asarray([dc, dr])))
        continue

      beams_prime.append((pos_prime, dir))

    beams = beams_prime

  return np.sum(np.max(energized, axis=2), dtype=int)

# Part 1
count_energized_tiles(arr, initial_beams)
```

```python
#initial_beams = [
#  (np.asarray([0, -1]), np.asarray([0, 1])),
#]

left_edge = [
  ((np.asarray([i, -1])), np.asarray([0, 1])) for i in range(arr.shape[0])
]

right_edge = [
  ((np.asarray([i, arr.shape[1]])), np.asarray([0, -1])) for i in range(arr.shape[0])
]

top_edge = [
  ((np.asarray([-1, i])), np.asarray([1, 0])) for i in range(arr.shape[1])
]

bottom_edge = [
  ((np.asarray([arr.shape[0], i])), np.asarray([-1, 0])) for i in range(arr.shape[1])
]

edges = left_edge + right_edge + top_edge + bottom_edge

max([count_energized_tiles(arr, [e]) for e in edges])
```
