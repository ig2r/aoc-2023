```python
import numpy as np

# Inputs for part 1 of puzzle:

# Sample
# inputs = [
#   (7, 9),
#   (15, 40),
#   (30, 200),
# ]

# Puzzle input
# inputs = [
#   (40, 215),
#   (92, 1064),
#   (97, 1505),
#   (90, 1100),
# ]

# Inputs for part 2 of puzzle:

# Sample
# inputs = [
#   (71530, 940200),
# ]

# Puzzle input
inputs = [
  (40929790, 215106415051100),
]

# Given a total race duration 't' and a record distance 'd', the actual distance 'x' travelled
# after holding the button for duration t0 at the beginning of the race is:
#
# d_actual = (t - t0) * t0
#
# To solve the puzzle, we substitute d (the record we have to beat) for d_actual and get:
# 0 = -(t0 ^ 2) + t * t0 - d
#
# I.e., it's a quadratic equation with coefficients -1, t, -d. This equation has two solutions,
# t0_1 and t0_2. All integers t0 between t0_1 and t0_2 will beat the record.

# Coefficient for t0 ^ 2 in quadratic equation
a = -1

# Small term to nudge t0_1/t0_2 off exact integer solutions before rounding since we want to
# beat the existing record, not just match it
epsilon = 0.001

def ways_to_beat_record(t, d):
  z = np.sqrt(t * t - 4 * a * (-d))
  t0_1, t0_2 = (-t - z) / 2 * a, (-t + z) / 2 * a
  t0_min, t0_max = np.ceil(min(t0_1, t0_2) + epsilon), np.floor(max(t0_1, t0_2) - epsilon)
  return t0_max - t0_min + 1

ways = [ways_to_beat_record(*input) for input in inputs]
np.prod(ways)
```
